```python
# return the first element greater than 10 in `my_list`
my_list = [0, 2, 3, 12, 4, 21]

found = False
for item in my_list:
    print(item)
    if item >= 10 and not found:
        res = item
        found = True
print(res)
```

```python
for item in my_list:
    print(item)
    if item >= 10:
        res = item
        break
print(res)
```

```python
# the for-else syntax
for item in [0, 1, 2, 3, -1, 20]:
    print(item)
    if item >= 10:
        res = item
        break
else:
    res = -99
    
print(res)
```

```python
def f(a):
    a += 1 # int is passed by value 
    return a 

a = 10
f(a)  # under the hood, a is copied into another variable, passed to f
# changing a inside f does not affect the global `a`
print(a)
```

```python
def set_to_one(some_list):
    some_list[0] = 1 # lists are passed by reference
    
a = [0, 1, 2]
set_to_one(a)
print(a)
```

```python
def set_to_one(some_list):
    some_list = 1 # passed by reference, 
    # but python creates a new variable inside this scope, called some_list, 
    # that overshadows the global one
    
a = [0, 1, 2]
set_to_one(a)
print(a)
```

```python
class People:
    # init is a special method, starting and ending with two underscores
    def __init__(self, name, age):
        self.name = name
        self.age = age
        self.n_calls = 0
    
    # I can define methods that are not special:
    # the first argument is (for now) always self
    def introduce(self):  # just a single argument
        print(f"Hello, I'm {self.name} and I'm {self.age} years old.")
        self.n_calls += 1
        print(f"This method has been called {self.n_calls} times")
```

```python
sylvain = People("Sylvain", 26)
```

```python
print(sylvain.name, sylvain.age)
```

```python
sylvain.introduce()
```

```python
introduce(sylvain)  # this doesn't work
```

```python
class CallablePeople:
    # init is a special method, starting and ending with two underscores
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def __call__(self):
        print("Hello")
        
callable_sylvain = CallablePeople("Sylvain", 26)
callable_sylvain()
    
```

```python
callable_sylvain[0] # we'd need to define a "__getitem__"
```

```python
# defaultdict 
from collections import defaultdict

def creator():
    print("A new key was inserted")
    return 10

my_default = defaultdict(creator)  # creator is a function 
# that returns the value inserted when the dict is tested for a missing key 

my_default['missing key'] 
```

```python
my_default
```

```python
my_default["other key"] = 0
```
