## 6
> **Check the equation numerically**

```python
def equality_check(n, p):
    np.random.seed(10)
    y = np.random.uniform(-1, 1, (n, 1))
    X = np.random.normal(0, np.sqrt(5), (n, p))
    lam = 1e-5
    left_side = X.T@np.linalg.solve(X@X.T + lam*np.identity(n), y)
    
    start2 = time.time()
    right_side = np.linalg.solve(X.T@X + lam*np.identity(p), X.T@y)
    end2 = time.time()
    return print(np.allclose(left_side, right_side))

equality_check(2000, 100)
```

<a name="2.7"></a>
## 7 
> **Compare computation time**

```python
def time_check(n, p):
    np.random.seed(10)
    y = np.random.uniform(-1, 1, (n, 1))
    X = np.random.normal(0, np.sqrt(5), (n, p))
    lam = 1e-5
    
    start1 = time.time()
    left_side = X.T@np.linalg.solve(X@X.T + lam*np.identity(n), y)
    end1 = time.time()
    time_left = end1 - start1
    
    start2 = time.time()
    right_side = np.linalg.solve(X.T@X + lam*np.identity(p), X.T@y)
    end2 = time.time()
    time_right = end2 - start2
    
    return print(f"time_left = {time_left} , time_right = {time_right}")
```

```python
time_check(5000, 100)
```

```python
time_check(100, 5000)
```

```python
time_check(500, 500)
```

**Observation**: For n>>p, the computation on the right side is significantly faster, and inversely for p>>n the left side is significantly faster. When n=p then both sides have approximately the same speed.  
**Explanation**: The np.linalg.solve method will solve a sysyem with n equations and n variables on the left and p equations with p variables on the right, so that is the difference in computation time.

# Exercise 3 (Random matrix spectrum)

<a name="3.8"></a>
## 8 
> **Choose three non-Gaussian probability distributions, with mean 0 and variance 2, and write a function that takes as input n, p and the distribution name, and creates a matrix X $\in \mathbb{R^(n,p)}$ with entries generated (i.i.d.) according to this distribution.**

We choose:
1. A uniform distribution $U([-\sqrt6 , \sqrt6])$
2. A Laplace distribution $Laplace(0,1)$
3. A logistic distribution $logistic(0, \sqrt6/\pi)$

```python
def distrib_mat(distrib_name, n, p):
    if distrib_name == 'uniform':
        return np.random.uniform(-math.sqrt(6), math.sqrt(6), (n, p))
    
    elif distrib_name == 'laplace':
        return np.random.laplace(0, 1, (n, p))
    
    elif distrib_name == 'logistic':
        return np.random.logistic(0, math.sqrt(6) / math.pi, (n, p))
    
    else:
        return "Error in distribution name"
```

<a name="3.9"></a>
## 9 
> **Display on one single graph the singular values of $X$ for $n = 1000$, and $p \in (200, 500, 1000, 2000)$ for
the three distributions chosen.**

```python
np.random.seed(10)
styles = [':', '--', '-', 'dashdot']
fig, ax = plt.subplots(1, figsize=(10, 6))
for p, i in zip([200, 500, 1000, 2000], range(4)):
    
    uniform_sing = np.linalg.svd(distrib_mat("uniform", 1000, p), compute_uv = False)
    laplace_sing = np.linalg.svd(distrib_mat("laplace", 1000, p), compute_uv = False)
    logistic_sing = np.linalg.svd(distrib_mat("logistic", 1000, p), compute_uv = False)
    plt.plot(np.sort(uniform_sing), linestyle = styles[i], color = 'b', label = f"SV uniform p = {p}")
    plt.plot(np.sort(laplace_sing), linestyle = styles[i], color = 'r', label = f"SV laplace p = {p}")
    plt.plot(np.sort(logistic_sing), linestyle = styles[i], color = 'g', label = f"SV logistic p = {p}")

plt.ylabel("Singular values")
plt.xlabel("Index of eigenvalues")
plt.title("Singular values of random matrices of size (1000, p) following different distributions of mean 0 and variance 2")
plt.legend()

plt.show()
```

**We have 12 plots, but visually it looks like only 4. That is because for a given p, all 3 random distribution matrices have almost identical singular values.**

<a name="3.10"></a>
## 10
> **Display on one single graph the spectrum (i.e. the set of eigen values) of $X^TX/n$ for $n = 1000$, and
$p \in 200, 500, 1000, 2000$. Comment**

```python
np.random.seed(10)
styles = [':', '--', '-', 'dashdot']
fig, ax = plt.subplots(1, figsize=(10, 6))
for p, i in zip([200, 500, 1000, 2000], range(4)):
    
    A_uni = distrib_mat("uniform", 1000, p)
    A_laplace = distrib_mat("laplace", 1000, p)
    A_log = distrib_mat("logistic", 1000, p)
    
    eigen_uni, _ = np.linalg.eigh(A_uni.T@A_uni / 1000)
    eigen_laplace, _ = np.linalg.eigh(A_laplace.T@A_laplace / 1000)
    eigen_log, _ = np.linalg.eigh(A_laplace.T@A_laplace / 1000)
    
    plt.plot(eigen_uni, linestyle = styles[i], color = 'b', label = f"Uniform, p = {p}")
    plt.plot(eigen_laplace, linestyle = styles[i], color = 'r', label = f"Laplace p = {p}")
    plt.plot(eigen_log, linestyle = styles[i], color = 'g', label = f"Logistic p = {p}")

plt.ylabel("Eigenvalues divided by 1000")
plt.xlabel("Index of eigenvalues")
plt.title("Eigenvalues/1000 of A.T@A with A being random matrices of size (1000, p) following different distributions of mean 0 and variance 2")
plt.legend()

plt.show()
```

This plot shows us that for $p = 2000$, our matrices A have a nullity of at least 1000. This can be explained because:

We know that the number of nonzero singular values of A equals the rank of A. The rank of a matrix of dimensions 1000x2000 is less than or equal to 1000, and therefore $A^TA$ has a maximum of 1000 nonzero eigenvalues.  
**This means the dimension of $Ker(A^TA)$ greater than or equal to 1000. That is exactly what we see in this graph.**




# Exercice 4. (Power method)

<a name="4.11"></a>
## 11 & 12

```python
def power_method(v, n, p, t):
    np.random.seed(10)
    U = np.zeros((n, t))
    V = np.zeros((p, t))
    X = np.random.normal(0, np.sqrt(5), (n, p))
    for k in range(t):
        u = X@v / np.linalg.norm(X@v)
        v = X.T@u / np.linalg.norm(X.T@u)
        U[:, k] = u
        V[:, k] = v
        sv1 = np.linalg.norm(X@v)
    return u, v, U, V, sv1
```

```python
# Lets calculate the true SVD using np.linalg.svd
np.random.seed(10)
X = np.random.normal(0, np.sqrt(5), (1000, 200))
U, S, VT = np.linalg.svd(X, full_matrices = True)
V = VT.T
v1 = V[:, 0]
u1 = U[:, 0]
```

```python
#We add "power" so we know we are dealing with approximations using the power method
#we add the "1" to the name because we approximate the first/lead SV and the associated vectors

v_init = np.ones(200)
u1_power, v1_power, U1_power, V1_power, _ = power_method(v_init, 1000, 200, 100)
```

```python
# Now let's plot whether our approximations of u1_power and v1_power converge to u1 and v1
def distance_plot(v, u1, v1, t):
    """
    Plots the norm between the approximations of u1 and v1  using the power method and the actual values of u1 and v1
    after each iteration.
    
    Args:
        v (ndarray) : (p,) initialization vector v0.
        u1 (ndarray) : (n,) the true value of the leading left singular vector of X.
        v1 (ndarray) : (p,) the true value of the leading right singular vector of X.
    
    Returns:
        Plot of the distances between approximation and true values.
    """
    
    _, _, U1_power, V1_power, _ = power_method(v, 1000, 200, t)
    t = U1_power.shape[1]
    distance_u = np.zeros(t)
    distance_v = np.zeros(t)
    for i in range(t):
        distance_u[i] = np.linalg.norm(U1_power[:, i] - u1)
        distance_v[i] = np.linalg.norm(V1_power[:, i] - v1)
    
    fig, ax = plt.subplots(2, figsize = (10, 6))
    ax[0].plot(distance_v)
    ax[1].plot(distance_u)
    
    ax[0].set_ylabel("Distance")
    ax[0].set_xlabel("Number of iterations")
    ax[1].set_ylabel("Distance")
    ax[1].set_xlabel("Number of iterations")
    
    ax[0].set_title("Distance after x iterations between power approx of v1 and true v1")
    ax[1].set_title("Distance after x iterations between power approx of u1 and true u1")
    
    plt.tight_layout()
```

```python
v_init = np.ones(200)
distance_plot(v_init, u1, v1, 100)
```

## 13
> **Provide two initialization vectors $v_0$ leading to different limits for this algorithm; explain how they
are related**
http://mlwiki.org/index.php/Power_Iteration

Mathematically we know that the column vectors $v_1,...,v_p$ of $V$ span $\mathbb{R^p}$. So our vector $v_0$ $\in \mathbb{R^p}$  can be represented as $v_0 =c_1v_1+...+c_pv_p$

The power method converges if and only if $v_0$ has a non zero component in the direction of the singular vector $v_1$, corresponding to the leading singular value. 

The power method converges if and only if **$c_1$ is non zero**.

Here we take two arrays of scalars $c = (c_1,..,cp)$, where the first array has the first element as 0 and the second array doesn't.

```python
scalars_non_cv = np.arange(200) #The important part is just that the first scalar is 0
scalars_cv =np.arange(200) + 1 #We add +1 because we want the first scalar to be non zero
```

```python
v_non_cv = V@scalars_non_cv
v_cv = V@scalars_cv 
```

```python
distance_plot(v_non_cv, u1, v1, 100)
```

```python
distance_plot(v_cv, u1, v1, 200)
```

## 14
> **Provide a way to approximate the largest singular value of X using the power method**

Mathematically we know that that norm of $ || Xv_1 ||  = \sigma_1 $.Therefore we use our approximation of $v1$ (called v1_power) to the approximate the singular value associated to it. Our results show that the approximation works.

```python
sv1_approx = np.linalg.norm(X@v1_power)
sv1 = S[0]
print(f"Our approximation of the leading singular value :{round(sv1_approx, 2)} and the actual value is:{round(S[0], 2)}")
```

## 15
> **Build upon the power method to provide an algorithm that can approximate the second largest
singular value of $X$ (without using an SVD function).**

The columns of the $V$ are the eigenvectors of the matrix $X^TX$. So through the first power method we got the leading singular value $\sigma_1$ of $X$, leading eigenvalue $\lambda_1 = \sigma_1^2$  of $A = X^TX$   and the leading eigenvector $v1$ associated to $A$., we do the following operation:

$A_2 = Aâˆ’Î»_1q_1q_1^T$

And then we redo the power method again, giving us $v2$, the biggest eigenvector associated with $A_2$, and the second biggest eigenvector associated with $A$, so the second leading singular vector. Once we have $v2$ we just need to calculate  
$ || Xv_2 ||  = \sigma_2 $

```python
def power_method_2(v, t):    
    u1_power, v1_power, _, _, sv1_power = power_method(v, 1000, 200, t)
    np.random.seed(10)
    X = np.random.normal(0,np.sqrt(5), (1000, 200))
    A2 = X.T@X - (sv1_power)**2 * np.outer(v1_power, v1_power)
    
    for k in range(t):
        z = A2@v
        v = z / np.linalg.norm(z)    
    sv2 = np.linalg.norm(X@v)    
    return sv2
```

```python
v_init = np.ones(200)
sv2_approx = power_method_2(v_init, 200)
print(f"Our approximation of the second largest singular value :{round(sv2_approx, 2)} and the actual value is: {round(S[1], 2)}")
```

It works!
