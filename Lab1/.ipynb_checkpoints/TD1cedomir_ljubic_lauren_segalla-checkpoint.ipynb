**Exercice 4** 
> Count the number of occurrences of each character in the string s="HelLo WorLd!!". (in real
life, you should use a collections.Counter ; here, you are asked to code the method yourself).
Output a dictionary that to each character associates the number of occurrences in this string. In
this question, we consider that lower and upper case characters are the same (e.g. your dictionary
should note have both a L and a l entry).


```python
test_str = "HelLo WorLd!!"
dict = {}
for letter in test_str.lower():
    counter = test_str.lower().count(letter)
    dict_temp = {letter : counter}
    dict = {**dict, **dict_temp} #in order to concatenate dictionnairies
print(dict)
```

> Compute 0.1 + 100 - 100. Using ==, check if it is equal to 0.1. Comment. Compare the two
floating point numbers again with an appropriate numpy function.

```python
x = 0.1 + 100 - 100
x == 0.1
```

```python
print('%10.25f' %(0.1))
```

```python
print('%10.25f' %(0.1 + 100 - 100))
```

We see that both values are not stored as exactly equal to 0.1 according to python. 

That is because floating-point numbers are represented in computer hardware as base 2 (binary) fractions. In other words, 0.1 gets converted from base 10 to base 2, where it is written as 0.000110011…(0011). 

Unfortunately the resulting binary number may not accurately represent the original base 10 number. 0.1 is one example. 

A consequence is that, in general, the decimal floating-point numbers we enter are only approximated by the binary floating-point numbers actually stored in the machine.

```python
np.allclose(x,0.1)
```

**Exercice 8** 

**Version 1**

```python
start1 = time.time()
prodv1 = 2
for k in range(1,10000):
    prodv1 = prodv1*(4*k**2/(4*k**2-1))
end1 = time.time()
print(f"The for loop took approx {round(end1 - start1,3)} seconds to compute")
```

**Version 2**

```python
start2 = time.time()

prodv2 = 2*np.prod(4*np.arange(1,10001)**2/(4*np.arange(1,10001)**2-1))

end2 = time.time()
end2 - start2

print(f"The numpy product took approx {round(end2 - start2,3)} seconds to compute")
```

```python
%%timeit
prodv2 = 2*np.prod(4*np.arange(1,10001)**2/(4*np.arange(1,10001)**2-1))
```

%%timeit runs the time calculation more than once (7 runs of 100 loops) and gives an average and stdv. So it is more precice and reliable than time.time()

> Create a random matrix M(5,6) with coefficients taken uniformly (and independently) in (-1,1).
Substract to each even column of M (say M[:, 0] is even), twice the value of the following (uneven)
column.

```python
M = np.random.uniform(-1,1,30).reshape(5, 6)
for i in range(M.shape[1]):
    if i%2 == 0:
        M[:,i] = M[:,i] - 2*M[:,i+1]
M
```

<a name="13"></a>
**Exercice 13** 


```python
np.random.seed(10)
M = np.random.uniform(-1,1,50).reshape(5, 10)
G = np.matmul(M.T,M)
G
```

```python
np.all(np.linalg.eigvals(G)>= -1e-15)
```

All eigenvalues are positive even though they are not technically positive according to python, which agasin has to do with how floats are saved in binary. With the matrix being symmetric and all the eigenvalues being positive, according to class we have that the matrix G is symmetric semi-definite positive. However it can not be said that 0 is not an eigenvalue

```python
np.linalg.matrix_rank(G)
```

```python
np.linalg.norm(G)
```

```python
np.linalg.norm(G,ord=2)
```

```python
np.std(G,axis=0)
```

**Exercice 14** 

```python
fig, ax = plt.subplots(1,1, figsize=(8,4))
x = np.linspace(-1,2,200)
y1 = x**2
y2 = x**3
y3 = x**4

plt.title("$x \mapsto x^d$ for d $\in$ {2,3,4}")
plt.xlabel("x")
plt.ylabel("y")
plt.plot(x,y1,label= r"$f(x) = x^{2}}$")
plt.plot(x,y2,label=r"$f(x) = x^{3}}$")
plt.plot(x,y3,label=r"$f(x) = x^{4}}$")
plt.legend(loc='upper left')
plt.show()
```
